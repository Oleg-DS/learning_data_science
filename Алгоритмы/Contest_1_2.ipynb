```python
import math as m

a = float(input())
b = float(input())
c = float(input())

try:
    d = m.pow(b, 2)-4*a*c
    print(m.sqrt(d))
    if d>0:
        x1 = (-b+m.sqrt(d))/(2*a)
        x2 = (-b-m.sqrt(d))/(2*a)
        if x1>x2:
            print(2, x2, x1)
        else:
            print(2, x1, x2)
    if d==0:
        x = (-b)/(2*a)
        print(1, x)
    if d<0:
        print(0)
except:
    print(3)
```

```python
n = input()

if '.' in n:
    res = 0
    for num in list(map(int, n.split('.'))):
        res = res<<8|num
    print(res)

else:
    n = int(n)
    print(f'{n>>24&255}.{n>>16&255}.{n>>8&255}.{n&255}')
```

```python
n = input()
if n.find('.') != -1:
    lst = list(map(int, n.split('.')))[::-1]
    print(lst[3]*256**3 + lst[2]*256**2 + lst[1]*256**1 + lst[0])
else:
    n = int(n)
    if n == 0:
        print('0.0.0.0')
    lst = []
    cnt = 0
    while n>0:
        lst.append(n%256)
        n = n // 256
        cnt += 1
    if cnt == 3:
        lst.append(0)
    elif cnt == 2:
        lst = lst + [0, 0]
    elif cnt == 1:
        lst = lst + [0, 0, 0]
    lst = list(map(str, lst[::-1]))
    print('.'.join(lst))
```

```python
base = float(input())
power = int(input())

result = 1
while power:
    if power&1 == 1:
        result = result*base
    base *= base
    power >>= 1
print(result)
```

```python
a = float(input())
n = int(input())

def pwr(x, y):
    if y == 0:
        return 1
    elif y%2 != 0:
        return x*pwr(x, (y-1))
    elif y%2 == 0:
        return pwr(x**2, y//2)

pwr(a, n)
```

```python
num = int(input())
bit_len = num.bit_length()
max_num = num
for _ in range(bit_len-1):
    num = (num>>1)|((num&1)<<(bit_len-1))
    if num > max_num:
        max_num = num
print(max_num)
```
