## Задание 2

1. Сгруппируйте набор данных babies по году и полу и сохраните результаты в два новых датафрейма: babies_girls и babies_boys.
2. Создайте фигуру matplotlib с 3 графиками один под другим.
3. Постройте линейные графики. Первый график должен показывать тренд рождаемости для девочек, второй - для мальчиков, третий объединять их все вместе (с теми же цветами, что в индивидуальных графиках). Годы - x, количество детей - y. 
4. Верхняя и правая границы графиков должны быть невидимы, к каждому графику должен быть заголовок, третий график должен содержать легенду, шкалы графиков должны быть подписаны.
5. Для шкалы количество должны быть установлены лимиты, чтобы она была одинакова на обоих графиках.
6. Кратко опишите тренды в ячейке markdown под графиками.

Если при группировке вы сделали год индексом, то можно обратиться к значениям этой переменной через аттрибут .index


```python
# ( ͡° ͜ʖ ͡°)づ ━━ ✫・*。

babies_girls = pd.DataFrame(b[(b.sex == 'F')].groupby('year')['number'].sum())
babies_boys = pd.DataFrame(b[(b.sex == 'M')].groupby('year')['number'].sum())

fig, ax = plt.subplots(3, figsize=(14, 10), gridspec_kw={'hspace':0.5})

for i in range(3):
    ax[i].set_xlabel('year')
    ax[i].spines['right'].set_visible(False)
    ax[i].spines['top'].set_visible(False)
    ax[i].set_ylim(top=2300000)

ax[0].plot(babies_girls.index, babies_girls.number, color='orange')
ax[0].set_ylabel('girls born')
ax[0].set_title('Number of girls born')

ax[1].plot(babies_boys.index, babies_boys.number, color='blue')
ax[1].set_ylabel('boys born')
ax[1].set_title('Number of boys born')

ax[2].plot(babies_girls.index, babies_girls.number, color='orange', label='girls born')
ax[2].plot(babies_boys.index, babies_boys.number, color='blue', label='boys born')
ax[2].set_ylabel('children born')
ax[2].legend(frameon=False)
ax[2].set_title('Number of children born')
```

### Краткие выводы о количестве новорожденных в США
* Тренды рождений для мальчиков и девочек, в целом, совпадают. Но если до 1930-ых гг. рождалось чуть больше девочек, то после 1940-ых гг. стало рождаться боьше мальчиков.
* С 1880-ых гг до 1950-ых гг. рождаемось детей росла, пик пришелся на 1950-1960-е гг.
* С 1960-ых гг. количество рождений находилось в диапазоне примерно 1500-2000 тыс. новорожденных.

## Задание 4

1. В оригинальном датафрейме babies создайте новую колонку - первая буква имени.
2. Выберете год из датасета. Сгруппируйте датасет, чтобы в нем в рядах были первые буквы, а в колонках - количество детей с такими именами. Сохраните три новых датафрейма для любых трех лет из выборки с такой группировкой.
3. Создайте фигуру matplotlib с 3 горизонтальными графиками один под другим.
4. Верхняя и правая границы графиков должны быть невидимы, каждый график быть с заголовком, шкалы графиков должны быть подписаны.
5. Постройте столбчатую диаграмму для каждого года. 
6. Сделайте вывод - какие первые буквы имени были самыми популярными в каждом году.

```python
# ( ͡° ͜ʖ ͡°)づ ━━ ✫・*。

babies['letter'] = babies['name'].apply(lambda x: str(x)[0])
b = babies.copy()
b1900 = pd.DataFrame(b[(b.year==1900)].groupby('letter')['number'].sum())
b1950 = pd.DataFrame(b[(b.year==1950)].groupby('letter')['number'].sum())
b2000 = pd.DataFrame(b[(b.year==2000)].groupby('letter')['number'].sum())
```

```python
fig, ax = plt.subplots(3, figsize=(14, 10), gridspec_kw={'hspace':0.7})

for i in range(3):
    ax[i].set_xlabel('name\'s first letter')
    ax[i].set_ylabel('number of children')
    ax[i].spines['right'].set_visible(False)
    ax[i].spines['top'].set_visible(False)

ax[0].bar(b1900.index, b1900.number)
ax[0].set_title('Number of children born in 1900 by the first letter of their name')

ax[1].bar(b1950.index, b1950.number)
ax[1].set_title('Number of children born in 1950 by the first letter of their name')

ax[2].bar(b2000.index, b2000.number)
ax[2].set_title('Number of children born in 2000 by the first letter of their name')
```

### Самые популярные первые буквы имен новорожденных:
* В 1900 году самой популярной была буква M
* В 1950 году самой популярной была буква J
* В 2000 году самой популярной была буква J

## Задание 5

1. Создайте фигуру matplotlib с двумя осями координат (1 ряд, две колонки)
2. В первой оси координат для датасета pima постройте мультивариативный график рассеяния. Шкала x - уровень глюкозы, шкала y - давление, размер - возраст, цвет - наличие диабета (Class). 
3. Во второй оси координат постройте мультивариативный график, где по x - количество беременностей, y - BMI, цвет - наличие диабета. У этого графика принудительно приведите значения шкалы x к дискретным (с помощью метода оси координат, смотрели такой для леса).
4. Верхняя и правая границы графиков должны быть невидимы, каждый график быть с заголовком, шкалы графиков должны быть подписаны.
5. По графикам вывод как эти переменные могут быть связаны с зависимой переменной (класс).


```python
# ( ͡° ͜ʖ ͡°)づ ━━ ✫・*。

p = pima.copy()

fig, ax = plt.subplots(1, 2, figsize=(20, 7))

for i in range(2):
    ax[i].spines['right'].set_visible(False)
    ax[i].spines['top'].set_visible(False)

cg0 = ax[0].scatter(p.Glucose, p.BloodPressure, s=p.Age*2, alpha=0.5, c=p.Class, cmap=plt.cm.get_cmap('viridis', 2))
ax[0].set_xlabel('glucose')
ax[0].set_ylabel('blood pressure')
ax[0].set_title('How blood pressure depends on glucose')
# plt.colorbar(cg0, ax=ax[0])

cg1 = ax[1].scatter(p.Pregnancies, p.BMI, s=p.Age*2, alpha=0.5, c=p.Class, cmap=plt.cm.get_cmap('viridis', 2))
ax[1].set_xlabel('number of pregnancies')
ax[1].set_ylabel('bmi')
ax[1].set_title('How BMI depends on a number of pregnancies')
ax[1].locator_params(integer=True)

plt.colorbar(cg1, ax=ax.ravel().tolist(), ticks=range(2))
```
